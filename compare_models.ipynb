```python
# BasicModelRunner doesn't exist in llama_index
# Use Hugging Face transformers instead
%pip install transformers torch
from transformers import AutoTokenizer, AutoModelForCausalLM
```

NON-FINETUNED MODELS

```python
# Load model and tokenizer (using smaller model for faster testing)
model_name = "microsoft/DialoGPT-small"  # Much smaller model (~117MB)
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Generate response
def generate_response(prompt, model, tokenizer):
    inputs = tokenizer.encode(prompt, return_tensors="pt")
    outputs = model.generate(inputs, max_length=100, num_return_sequences=1)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

non_finetuned_output = generate_response("Tell me how to train my dog to sit", model, tokenizer)
```

```python
print(non_finetuned_output)
```

FINE-TUNED MODELS

```python
# Load fine-tuned model (using smaller model for faster testing)
finetuned_model_name = "microsoft/DialoGPT-medium"  # Medium size model (~345MB)
finetuned_tokenizer = AutoTokenizer.from_pretrained(finetuned_model_name)
finetuned_model = AutoModelForCausalLM.from_pretrained(finetuned_model_name)
```

```python
finetuned_output = generate_response("Tell me how to train my dog to sit", finetuned_model, finetuned_tokenizer)
```

```python
print(finetuned_output)
```
